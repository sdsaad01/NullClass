```python
# first of all we will import all the libraries
import pandas as pd
import plotly.express as px
from datetime import datetime
import pytz
```

```python
# now we will take the data of the file which is provided to us
df = pd.read_csv("C:/Users/SYED SAAD/Downloads/Play Store Data.csv")
```

```python

# first we will work on the install column 
df['Installs'] = df['Installs'].str.replace('+', '', regex=False).str.replace(',', '', regex=False)
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')
df = df.dropna(subset=['Installs'])
```

```python
df.head()
```

```python

# Filter out categories starting with A, C, G, S as per given in the task
df = df[~df['Category'].str.startswith(('A', 'C', 'G', 'S'))]

```

```python
# Group by category and sum installs
category_installs = df.groupby('Category')['Installs'].sum().reset_index()


```

```python
# Get top 5 categories
top_categories = category_installs.sort_values(by='Installs', ascending=False).head(5)

```

```python
# Add mock country mapping (for demonstration)
countries = ['US', 'IN', 'BR', 'GB', 'DE']
top_categories['Country'] = countries
top_categories['Highlight'] = top_categories['Installs'] > 1_000_000

```

```python
# Check IST time
ist = pytz.timezone('Asia/Kolkata')
current_time = datetime.now(ist)
current_hour = current_time.hour
```

```python

# Show plot only between 6 PM and 8 PM IST
if 18 <= current_hour < 20:
    fig = px.choropleth(
        top_categories,
        locations='Country',
        color='Installs',
        hover_name='Category',
        color_continuous_scale='Viridis',
        title='Global Installs by Top 5 Categories (Filtered)'
    )
    fig.show()
else:
    print("The graph is not displayed because the current time is outside 6 PM - 8 PM IST.")
```
