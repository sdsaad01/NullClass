```python
import pandas as pd
```

```python
# getting my csc users report here
df = pd.read_csv('C:/Users/SYED SAAD/Downloads/Play Store Data.csv')
df.head()
```

```python
# Step 1: Convert to string (if not already)
df['Size'] = df['Size'].astype(str)

# Step 2: Remove 'M' and handle 'k' or other characters if needed
df['Size'] = df['Size'].str.replace('M', '', regex=False)

# Optional: You can also filter out "Varies with device" or other non-numeric sizes
df['Size'] = pd.to_numeric(df['Size'], errors='coerce')
```

```python
df.head()
```

```python
# Step: Clean 'Installs' column
# Convert to string to allow .str operations
df['Installs'] = df['Installs'].astype(str)  
 # Remove + and ,
df['Installs'] = df['Installs'].str.replace('[+,]', '', regex=True) 
# Convert to numeric
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')  
```

```python
df.head()
```

```python
# cleaning the reviews column
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
```

```python
df.head()
```

```python
df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce')
```

```python
df.head()
```

```python
#filtering the data on the basis of ratings 
filtered_df = df[
    (df['Rating'] >= 4.0) &
    (df['Size'] >= 10) &
    (df['Last Updated'].dt.month == 1)
]
#here we had made different sections on the basis of ratings
```

```python
#now we will try to get top 10 categories by installs number
top_categories = (
    filtered_df.groupby('Category')['Installs']
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .index
)

top_df = filtered_df[filtered_df['Category'].isin(top_categories)]
```

```python
# Aggregate Average Rating and Total Reviews
agg_df = top_df.groupby('Category').agg({
    'Rating': 'mean',
    'Reviews': 'sum'
}).reset_index()
```

```python
# now our second most important thing starts from here that we have to plot the graphs so we will install labraries
# we will be using bar graph
import matplotlib.pyplot as plt
import numpy as np
```

```python
#writing code here on the basis of category , rating and reviews we will do

labels = agg_df['Category']
ratings = agg_df['Rating']
reviews = agg_df['Reviews']
#providing the dimensions to our graph
x = np.arange(len(labels))
width = 0.35

plt.figure(figsize=(12, 6))
plt.bar(x - width/2, ratings, width, label='Avg Rating')
plt.bar(x + width/2, reviews, width, label='Total Reviews')

plt.xlabel('App Category')
plt.ylabel('Value')
plt.title('Top 10 App Categories: Avg Rating vs Total Reviews')
plt.xticks(x, labels, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
